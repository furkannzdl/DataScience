```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("/Users/furkanozdal/Downloads/UNZIP_FOR_NOTEBOOKS_FINAL/08-Linear-Regression-Models/Advertising.csv")
```

```python
X = df.drop("sales",axis=1)
y = df["sales"]
```

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
polynomial_converter = PolynomialFeatures(include_bias=False,degree=3)
```

```python
poly_features = polynomial_converter.fit_transform(X)
```

```python
X.shape
```

```python
poly_features.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(poly_features, y, test_size=0.3, random_state=101)
```

```python
X_train.shape
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
scaler.fit(X_train)
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

```python
X_test
```

```python
from sklearn.linear_model import Ridge
```

```python
ridge_model = Ridge(alpha=10)
```

```python
ridge_model.fit(X_train, y_train)
```

```python
test_pred = ridge_model.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error
```

```python
mean_absolute_error(y_test,test_pred)
```

```python
np.sqrt(mean_squared_error(y_test,test_pred))
```

```python
from sklearn.linear_model import RidgeCV
```

```python
ridge_cv_model = RidgeCV(alphas=(0.1, 1.0, 10.0))
```

```python
ridge_cv_model.fit(X_train,y_train)
```

```python
ridge_cv_model.alpha_
```

```python
from sklearn.linear_model import LassoCV
```

```python
lasso_cv_model = LassoCV(max_iter=1000000)
```

```python
lasso_cv_model.fit(X_train,y_train)
```

```python
lasso_cv_model.alpha_
```

```python
test_predictions = lasso_cv_model.predict(X_test)
```

```python
mean_absolute_error(y_test,test_predictions)
```

```python
np.sqrt(mean_squared_error(y_test,test_predictions))
```

```python
lasso_cv_model.coef_
```

```python
from sklearn.linear_model import ElasticNetCV
```

```python
elastic_model = ElasticNetCV(l1_ratio=[.1, .5, .7, .9, .95, .99, 1],max_iter=1000000)
```

```python
elastic_model.fit(X_train,y_train)
```

```python
elastic_model.l1_ratio_
```

```python
test_pred = elastic_model.predict(X_test)
```

```python
mean_absolute_error(y_test,test_pred)
```

```python
np.sqrt(mean_squared_error(y_test,test_pred))
```
