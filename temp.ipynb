```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def create_ages(mu=50, sigma=13, num_sample=100, seed=42):
    np.random.seed(seed)
    sample_ages = np.random.normal(loc=mu,scale=sigma,size=num_sample)
    sample_ages = np.round(sample_ages,decimals=0)

    return sample_ages
```

```python
sample = create_ages()
```

```python
sns.histplot(sample,bins=30,kde=True)
```

```python
sns.boxplot(sample)
```
